## Train-test split

```python
np.random.seed(7)
molecules_names = pd.unique(raw_data[0])
n_molecules = len(molecules_names)

train_molecules = np.random.choice(molecules_names, int(n_molecules * 0.9), replace=False)
train_idx = np.in1d(raw_data[0], train_molecules)

train_data = raw_data[train_idx]
test_data = raw_data[~train_idx]
```

```python
X_train = np.array(train_data.values[:, 2:-1], dtype=float)
y_train = np.array(train_data.values[:,-1], dtype=bool)

X_test = np.array(test_data.values[:, 2:-1], dtype=float)
y_test = np.array(test_data.values[:,-1], dtype=bool)
```

## K-fold split

```python
import xgboost as xgb
from sklearn.model_selection import KFold
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
```

```python
def get_train_test_data(train_names):
    
    train_idx = np.in1d(raw_data[0], train_names)

    train_data = raw_data[train_idx]
    test_data = raw_data[~train_idx]

    X_train = np.array(train_data.values[:, 2:-1], dtype=float)
    y_train = np.array(train_data.values[:,-1], dtype=bool)

    X_test = np.array(test_data.values[:, 2:-1], dtype=float)
    y_test = np.array(test_data.values[:,-1], dtype=bool)
    
    bags_train = np.array(train_data.values[:, 0], dtype=str)
    bags_test = np.array(test_data.values[:, 0], dtype=str)
    
    return X_train, X_test, y_train, y_test, bags_train, bags_test


def mil_score(true, pred, bags):
    
    for bag_name in np.unique(bags):
        idx = np.in1d(bags, bag_name)
        pred[idx] = np.max(pred[idx])
    
    return np.mean(true == pred)
```

```python
molecules_names = pd.unique(raw_data[0])
musk_names = molecules_names[map(lambda x: 'NON' not in x, molecules_names)]
non_musk_names = molecules_names[map(lambda x: 'NON' in x, molecules_names)]
```

```python
classifiers = [
    KNeighborsClassifier(n_neighbors=8, metric='manhattan'),
    RandomForestClassifier(n_estimators=50, max_features='log2'),
    MLPClassifier(hidden_layer_sizes=(300, 200,)),
    AdaBoostClassifier(),
    SVC(class_weight={1:5}, probability=True)
]
# parameters for all classifiers have been optimized via grid search
```

```python
kf = KFold(n_splits=10, shuffle=True)
np.random.seed(7)
classifiers_scores = []

for (musk_train_index, musk_test_index), (non_musk_train_index, non_musk_test_index) in \
                        zip(kf.split(musk_names), kf.split(non_musk_names)):

    train_names = np.hstack((musk_names[musk_train_index], non_musk_names[non_musk_train_index]))
    X_train, X_test, y_train, y_test, bags_train, bags_test = get_train_test_data(train_names)

    scores = []
    for clf in classifiers:
        clf.fit(X_train, y_train)
        pred = clf.predict(X_test)
        scores.append(mil_score(y_test, pred, bags_test))
        
    classifiers_scores.append(scores)
    
print(np.mean(classifiers_scores, axis=0))
```

```python
class StackedClassifier():
    """
    A stacking classifier.
    
    Parameters
    ----------
    bclf : stage1 model for stacking.
    clfs : list of stage0 machine learning models.
    """
    
    def __init__(self, bclf, clfs):
        self.bclf = bclf
        self.clfs = clfs


    def fit_base_clfs(self, X, y):
        for i in range(len(self.clfs)):
            self.clfs[i].fit(X, y)


    def predict_probas_base_clfs(self, X):
        probas = np.zeros((X.shape[0], len(self.clfs)))
        for i in range(len(self.clfs)):
            probas[:, i] = clf.predict_proba(X)[:,1]
            
        return probas


    def fit(self, X, y):
        self.fit_base_clfs(X, y)
        X = np.hstack((X, self.predict_probas_base_clfs(X)))
        self.bclf.fit(X, y)


    def predict_proba(self, X):
        """Predict class probabilities for X.
        The predicted class probabilities of an input sample is computed.
        
        Parameters
        ----------
        X : array-like or sparse matrix of shape = [n_samples, n_features]
            The input samples.
        Returns
        -------
        p : array of shape = [n_samples, n_classes].
            The class probabilities of the input samples.
        """
        X = np.hstack((X, self.predict_probas_base_clfs(X)))

        return self.bclf.predict_proba(X)


    def predict(self, X):
        """Predict class for X.
        The predicted class of an input sample is a vote by the StackedClassifier.
        Parameters
        ----------
        X : array-like or sparse matrix of shape = [n_samples, n_features]
            The input samples. Internally, it will be converted to
            ``dtype=np.float32`` and if a sparse matrix is provided
            to a sparse ``csr_matrix``.
        Returns
        -------
        y : array of shape = [n_samples]
            The predicted classes.
        """
        X = np.hstack((X, self.predict_probas_base_clfs(X)))

        return self.bclf.predict(X)


    def score(self, X, y):
        X = np.hstack((X, self.predict_probas_base_clfs(X)))
        
        return self.bclf.score(X, y)
```

```python
bclf = xgb.XGBClassifier()
model = StackedClassifier(bclf, classifiers)

model_scores = []
np.random.seed(7)
for (musk_train_index, musk_test_index), (non_musk_train_index, non_musk_test_index) in \
                        zip(kf.split(musk_names), kf.split(non_musk_names)):

    train_names = np.hstack((musk_names[musk_train_index], non_musk_names[non_musk_train_index]))
    X_train, X_test, y_train, y_test, bags_train, bags_test = get_train_test_data(train_names)

    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    
    bclf.fit(X_train, y_train)
    bclf_pred = bclf.predict(X_test)
    
    model_scores.append([mil_score(y_test, pred, bags_test), mil_score(y_test, bclf_pred, bags_test)])
    
print(np.mean(model_scores, axis=0))
```
